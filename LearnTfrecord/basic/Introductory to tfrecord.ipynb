```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# Convert an image (NumPy array) to bytes
def _bytes_feature(value):
    """Returns a bytes_list from a string/byte."""
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))

def _int64_feature(value):
    """Returns an int64_list from an int/boolean."""
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))

def serialize_example(image, label):
    """
    Creates a tf.train.Example message from image and label.
    """
    feature = {
        "image": _bytes_feature(image.tobytes()),  # Serialize image as bytes
        "label": _int64_feature(label),  # Serialize label as int
    }
    # Create a Features message using tf.train.Example
    example_proto = tf.train.Example(features=tf.train.Features(feature=feature))
    return example_proto.SerializeToString()
```

```python
def write_tfrecord(file_name, images, labels):
    with tf.io.TFRecordWriter(file_name) as writer:
        for image, label in zip(images, labels):
            serialized_example = serialize_example(image, label)
            writer.write(serialized_example)

# Save training and test datasets
write_tfrecord("mnist_train.tfrecord", x_train, y_train)
write_tfrecord("mnist_test.tfrecord", x_test, y_test)
print("TFRecord files created successfully!")
```

```python
def parse_example(example_proto):
    # Define the feature description to parse the TFRecord
    feature_description = {
        "image": tf.io.FixedLenFeature([], tf.string),
        "label": tf.io.FixedLenFeature([], tf.int64),
    }
    parsed_features = tf.io.parse_single_example(example_proto, feature_description)
    # Decode the image back to its original shape
    image = tf.io.decode_raw(parsed_features["image"], tf.uint8)
    image = tf.reshape(image, (28, 28))  # MNIST images are 28x28
    label = parsed_features["label"]
    return image, label

# Load TFRecord file
raw_dataset = tf.data.TFRecordDataset("mnist_train.tfrecord")
parsed_dataset = raw_dataset.map(parse_example)

# Check the first example
for image, label in parsed_dataset.take(1):
    print("Image shape:", image.shape)
    print("Label:", label.numpy())
```

```python
def parse_example(serialized_example):
    feature_description = {
        'image': tf.io.FixedLenFeature([], tf.string),  # Read as string (bytes)
        'label': tf.io.FixedLenFeature([], tf.int64)   # Read as integer
    }
    parsed_example = tf.io.parse_single_example(serialized_example, feature_description)

    # Decode the bytes back to the image format and reshape
    image = tf.io.decode_raw(parsed_example['image'], tf.uint8)  # Decode raw bytes
    image = tf.reshape(image, (28, 28, 1))  # Reshape to original dimensions
    image = tf.cast(image, tf.float32) / 255.0  # Normalize to [0, 1]

    label = parsed_example['label']
    return image, label

```

```python
# build complte pipeline

def prepare_dataset(tfrecord_file, batch_size=32, shuffle_buffer_size=1000):
    raw_dataset = tf.data.TFRecordDataset(tfrecord_file)
    parsed_dataset = raw_dataset.map(parse_example)
    dataset = (
        parsed_dataset
        .shuffle(shuffle_buffer_size)
        .batch(batch_size)
        .prefetch(tf.data.AUTOTUNE)  # Optimizes pipeline execution
    )
    return dataset

train_dataset = prepare_dataset("/content/mnist_train.tfrecord", batch_size=64)
test_dataset = prepare_dataset("/content/mnist_test.tfrecord", batch_size=64)
```

```python
for image, label in train_dataset.take(1):
    print("Image shape:", image.shape)
    print("Label:", label.numpy())
```

```python
import time
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
start=time.time()

model.fit(
    train_dataset,
    validation_data=test_dataset,
    epochs=10
)
end_time=time.time()
print(end_time-start)
```

```python
# now write gzip compressed tfrecord


def compressed_tfrecord(filename,images,labels):
  options=tf.io.TFRecordOptions(compression_type="GZIP")
  with tf.io.TFRecordWriter(filename,options) as writer:
    for image,label in zip(images,labels):
      serialized_example=serialize_example(image,label)
      writer.write(serialized_example)

compressed_tfrecord("mnist_train_compressed.tfrecord", x_train, y_train)
compressed_tfrecord("mnist_test_compressed.tfrecord", x_test, y_test)

def prepare_compressed_dataset(filename, batch_size):
    raw_dataset = tf.data.TFRecordDataset(
        filename,
        compression_type="GZIP"  # Specify the compression type
    )
    parsed_dataset = raw_dataset.map(parse_example)
    dataset = parsed_dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return dataset


train_dataset_gzip = prepare_compressed_dataset("/content/mnist_train_compressed.tfrecord", batch_size=64)
test_datasetgzip = prepare_compressed_dataset("/content/mnist_test_compressed.tfrecord", batch_size=64)
```
